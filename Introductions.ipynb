# Class Summary
We will focus on social media data, and how insights and methods from Psychology and Cognitive Science can help us make sense of the complex datasets that are generated by human behavior on large-scale social media platforms. Part one will focus on [Misinformation on Twitter](https://www.nature.com/articles/s41586-024-07524-8). Part Two will focus on [real world decision-making on Reddit](https://osf.io/preprints/psyarxiv/hzv7c). We will analyze datasets using Python.
